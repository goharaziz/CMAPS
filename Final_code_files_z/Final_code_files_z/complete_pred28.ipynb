```python
tr_df.columns=['posting', 'date', 'month', 'year', 'day', 'time', 'offences', 'crime_categories','partoftheday','latitude','longitude','location']
```

```python
# new data frame with split value columns 
new = tr_df['time'].str.split(":", n = 1, expand = True) 
  
# making separate first name column from new data frame 
tr_df["hour"]= new[0] 
  
# making separate last name column from new data frame 
tr_df["minute"]= new[1] 
  
# Dropping old Name columns 
tr_df.drop(columns =["time"], inplace = True) 
  
# df display 
tr_df 
```

```python
z_df1 = tr_df.loc[tr_df['crime_categories'].isin(['Crime Against Property','Crime Against Person', 'Ordinance','Police Acts'])]
z_df1.head()
```

```python
z_df1.loc[z_df1['posting'].isin(['Sarwar Road'])]
```

```python
# Sort the rows of dataframe by column 
z_df1 = z_df1.sort_values(by ='crime_categories')

z_df1.head()
```

```python
z_df1['crime_categories']= label_encoder.fit_transform(z_df1['crime_categories']) 
z_df1['crime_categories'].unique()
```

```python
z_df1.head()
```

# key 'Crime Against Person','Crime Against Property', 'Ordinance','Police Acts' = 0,1,2,3

```python
z_df1['day']= label_encoder.fit_transform(z_df1['day']) 
z_df1['day'].unique()
z_df1['posting']= label_encoder.fit_transform(z_df1['posting']) 
z_df1['posting'].unique()
z_df1['partoftheday']= label_encoder.fit_transform(z_df1['partoftheday']) 
z_df1['partoftheday'].unique()
z_df1['location']= label_encoder.fit_transform(z_df1['location']) 
z_df1['location'].unique()
```

```python
X = z_df1.drop([ 'offences', 'crime_categories'] , axis=1)
X.head()
```

```python
y = z_df1.drop(['posting', 'date', 'month', 'year', 'day', 'hour','minute', 'offences','partoftheday','latitude','longitude','location'] , axis=1)
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```
