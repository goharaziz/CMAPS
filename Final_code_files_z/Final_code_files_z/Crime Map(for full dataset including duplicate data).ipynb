```python
unique_locations = df['TGAT'].value_counts()
```

```python
CR_index = pd.DataFrame({"Raw_String" : unique_locations.index, "ValueCount":unique_locations})
CR_index.index = range(len(unique_locations))
CR_index.head()
```

```python
def Location_extractor(Raw_Str):
    preProcess = Raw_Str.split(',')
    lat =  float(preProcess[0])
    long = float(preProcess[1])
    return (lat, long)
```

```python
CR_index['LocationCoord'] = CR_index['Raw_String'].apply(Location_extractor)
```

```python
CR_index  = CR_index.drop(columns=['Raw_String'], axis = 1)
```

```python
%%time

lahore_map_crime = folium.Map(location=[31.5204,74.35874],
                        zoom_start=13,
                        )

for i in range(len(CR_index)):
    lat = CR_index['LocationCoord'].iloc[i][0]
    long = CR_index['LocationCoord'].iloc[i][1]
    radius = CR_index['ValueCount'].iloc[i] / 45
    
    if CR_index['ValueCount'].iloc[i] > 1000:
        color = "#FF0000"
    elif CR_index['ValueCount'].iloc[i] > 800:
         color = "#FF4500"
    elif CR_index['ValueCount'].iloc[i] > 500:
         color = "#32CD32"
    elif CR_index['ValueCount'].iloc[i] > 200:
         color = "#FFFF00"
    else:
        color = "#008080" 
    z=range(len(Crime_type1))
    
    popup_text ="""
                Area : {}<br>
                Crime Incidents: {} <br>"""
    popup_text = popup_text.format(area['Area'].iloc[i],
                               CR_index['ValueCount'].iloc[i]
                               )
    folium.CircleMarker(location = [lat, long], popup= popup_text,radius = radius, color = color, fill = True).add_to(lahore_map_crime)
```
